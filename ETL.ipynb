# Import

```python
import pandas as pd
import datetime
import fastparquet
```

# Uploaded "flights.csv", applied a filter (date from April 2020 till now), created new dataframe with the filter

```python
df_flights = pd.read_csv("flights.csv")

condition_date = datetime.datetime(2020, 4, 1)
current_date = datetime.datetime.today()

df_flights["date"] = pd.to_datetime(df_flights["date"])
mask = (df_flights["date"] >= condition_date) & (df_flights["date"] <= current_date)
new_df_flights = df_flights.loc[mask]

new_df_flights
```

# Uploaded "users.csv", applied a filter (gender only female and age older than 35 years), created new dataframe with the filter

```python
df_users = pd.read_csv("users.csv")
new_df_users = df_users.loc[(df_users["gender"] == "female") & (df_users["age"] > 35)]

new_df_users
```

# Uploaded "hotels.csv"

```python
df_hotels = pd.read_csv("hotels.csv")

df_hotels
```

```python
joined_flights_hotels = pd.merge(new_df_flights, df_hotels, how = "inner", on = "travelCode")
joined_all_df = pd.merge(new_df_users, joined_flights_hotels, how = "inner", left_on = "code", right_on = "userCode_x")

joined_all_df
```
